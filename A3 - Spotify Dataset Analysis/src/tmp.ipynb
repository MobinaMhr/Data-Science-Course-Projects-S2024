```python
df = spark.read.parquet(SPOTIFY)

df.show()
```

```python
df.printSchema()
```

```python
df = df.withColumn("release_date", to_date(df["release_date"]))

df.show()
```

```python
for col in df.columns:
    print(f"{col}: {df.filter(df[col].isNull()).count()}")
```

```python
df = df.na.drop()

for col in df.columns:
    print(f"{col}: {df.filter(df[col].isNull()).count()}")
```

```python
df = df.withColumn("song_age", lit(2024) - year("release_date"))

df.show()
```

```python
assembler = VectorAssembler(inputCols=["loudness"], outputCol="loudness_vec")
df = assembler.transform(df)

scaler = StandardScaler(inputCol="loudness_vec", outputCol="loudness_zscore", withStd=True, withMean=True)

scalerModel = scaler.fit(df)

df = scalerModel.transform(df)

first_element = udf(lambda v: float(v[0]), DoubleType())
df = df.withColumn("loudness_zscore", first_element("loudness_zscore"))

df = df.filter(abs(df["loudness_zscore"]) < 3)
```
